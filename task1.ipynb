# project description
- create a barchart or histogram to visualize the distribution of a categorical or continuous variable.
- world population dataset is downloaded from kaggle site(https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset?resource=download)
- im looking at continental data vs population with respect to the years.

```python
# firstly required libraries are imported 
import pandas as pd   # to work on csv file
import matplotlib.pyplot as plt  # used for data visualization 

```

```python
# read the csv file
df=pd.read_csv(r"C:\Users\Asus\OneDrive\Desktop\DataScienceTasks\task1\data\world_population.csv")
```

```python
# exploring the dataset
print("shape of the dataset:",df.shape)
```

```python
print(df.info())
```

```python
# columns of dataset
print(df.columns)          
```

```python
# plotting the dataset.
x=df["Continent"]
y=df["2022 Population"]
plt.bar(x,y,color='navy',width=0.6,alpha=0.5)   # edgecolor="black"???
plt.xlabel("Continent",fontsize=15)
plt.ylabel("Population(in billions)",fontsize=15)
plt.title("POPULATION IN DIFFERENT CONTIENENTS",fontsize=15)
plt.xticks(rotation=30)
plt.show()
```

```python
# continental data versus population data with respect to year.
x=df["Continent"]
y1=df["2022 Population"]
y2=df["2020 Population"]
plt.bar(x,y1,width=0.25)
plt.bar(x,y2,width=0.5)
plt.show()
```

```python
continent_population=df.groupby("Continent").sum()
print(continent_population["2022 Population"])
```

```python
print(continent_population.columns)
```

```python
x=continent_population.index.tolist()
y=continent_population["2022 Population"].values
plt.bar(x,y,color="navy",alpha=0.75)
plt.xlabel("Continents",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.title("POPULATION IN DIFFERENT CONTINENTS")
plt.xticks(rotation=30)
plt.show()
```
